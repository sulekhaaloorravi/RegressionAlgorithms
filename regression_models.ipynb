### Gaussian Process

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# Create a synthetic dataset
# Input data (X) and corresponding target values (y)
X = np.linspace(0, 5, 100).reshape(-1, 1)  # 100 points between 0 and 5
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])  # Sinusoidal data with noise

# Define the kernel: Constant kernel times Radial Basis Function (RBF) kernel
kernel = C(1.0, (1e-4, 1e1)) * RBF(1.0, (1e-4, 1e1))

# Instantiate the Gaussian Process Regressor with the specified kernel
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=1e-2)

# Fit the model
gp.fit(X, y)

# Make predictions on test data
X_test = np.linspace(0, 5, 1000).reshape(-1, 1)  # Generate 1000 test points between 0 and 5
y_pred, sigma = gp.predict(X_test, return_std=True)  # Predict mean and standard deviation

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(X, y, 'r.', markersize=10, label="Training data")  # Training data points
plt.plot(X_test, y_pred, 'b-', label="Prediction mean")  # Prediction mean
plt.fill_between(X_test.ravel(), y_pred - 1.96 * sigma, y_pred + 1.96 * sigma, 
                 alpha=0.2, color='gray', label="95% confidence interval")  # Confidence interval
plt.title("Gaussian Process Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend(loc="upper left")
plt.show()
```

### ElasticNet

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Create a synthetic regression dataset
X, y = make_regression(n_samples=1000, n_features=20, noise=0.1, random_state=42)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instantiate the Elastic Net model
# alpha controls the regularization strength (higher values mean stronger regularization)
# l1_ratio controls the mix between Lasso (l1_ratio=1) and Ridge (l1_ratio=0) penalties
elastic_net = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)

# Fit the model on the training data
elastic_net.fit(X_train, y_train)

# Make predictions on the test data
y_pred = elastic_net.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Plot true vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color="blue", edgecolor="k", alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color="red", lw=2)
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("Elastic Net: True vs Predicted Values")
plt.show()
```

### Deep Learning

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Importing necessary components from Keras
from keras.models import Sequential
from keras.layers import Dense

# Step 1: Create a synthetic dataset for regression
X, y = make_regression(n_samples=1000, n_features=20, noise=10, random_state=42)

# Step 2: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Standardize the dataset (very important for neural networks)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 4: Build a simple feedforward neural network model using Keras
model = Sequential()

# Input layer + hidden layer 1 with 64 neurons and 'relu' activation
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))

# Hidden layer 2 with 64 neurons
model.add(Dense(64, activation='relu'))

# Output layer (single neuron for regression output)
model.add(Dense(1))

# Step 5: Compile the model
# Use 'adam' optimizer, mean squared error loss for regression, and track mean squared error
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])

# Step 6: Train the model on the training data
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)

# Step 7: Make predictions on the test set
y_pred = model.predict(X_test)

# Step 8: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Step 9: Plot true vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color="blue", edgecolor="k", alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color="red", lw=2)
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("Neural Network: True vs Predicted Values")
plt.show()

# Step 10: Plot training loss and validation loss
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.title('Training and Validation Loss over Epochs')
plt.legend()
plt.show()
```

### Polynomial Regression

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Step 1: Create a synthetic dataset
# We'll use a quadratic function with some noise
np.random.seed(42)
X = np.random.rand(100, 1) * 6 - 3  # 100 points between -3 and 3
y = 0.5 * X**2 + X + 2 + np.random.randn(100, 1) * 0.5  # Quadratic equation with noise

# Step 2: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Apply polynomial transformation to the features
# PolynomialFeatures transforms the original features into polynomial features
degree = 2  # We are applying quadratic transformation
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

# Step 4: Train a Linear Regression model on the transformed features
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Step 5: Make predictions on the test set
y_pred = model.predict(X_test_poly)

# Step 6: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Step 7: Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue', label='True values')
plt.scatter(X_test, y_pred, color='red', label='Predicted values', alpha=0.7)
plt.xlabel("X")
plt.ylabel("y")
plt.title(f"Polynomial Regression (Degree {degree})")
plt.legend()
plt.show()

# Step 8: Visualize the polynomial fit on the entire dataset
X_range = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
y_range_pred = model.predict(X_range_poly)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_range, y_range_pred, color='red', label='Polynomial Fit')
plt.xlabel("X")
plt.ylabel("y")
plt.title(f"Polynomial Regression Fit (Degree {degree})")
plt.legend()
plt.show()
```

### Quantum Machine Learning - I

```python
# Import necessary libraries
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

# Step 1: Create a synthetic dataset (binary classification)
# We'll use the 'make_moons' dataset for a simple classification task
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)
y = y * 2 - 1  # Convert labels to {-1, 1} for compatibility

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 2: Define a quantum device (2 qubits)
n_qubits = 2
dev = qml.device("default.qubit", wires=n_qubits)

# Step 3: Define a quantum circuit (variational circuit)
@qml.qnode(dev)
def quantum_circuit(weights, x):
    # Encode the input features into the quantum state
    qml.AngleEmbedding(x, wires=range(n_qubits))
    
    # Apply variational rotation gates
    qml.BasicEntanglerLayers(weights, wires=range(n_qubits))
    
    # Measure the expectation value of PauliZ for binary classification
    return qml.expval(qml.PauliZ(0))

# Step 4: Define the cost function
def cost(weights, X, y):
    # Predict for each example and compute the squared error
    predictions = np.array([quantum_circuit(weights, x) for x in X])
    return np.mean((predictions - y) ** 2)

# Step 5: Initialize parameters and set optimization
np.random.seed(42)
weights = np.random.randn(3, n_qubits)  # Initialize random weights

# Use the Adam optimizer for training
opt = qml.AdamOptimizer(stepsize=0.1)
epochs = 100

# Track the cost over epochs
costs = []

# Step 6: Train the model
for epoch in range(epochs):
    # Perform optimization step
    weights, cost_value = opt.step_and_cost(lambda w: cost(w, X_train, y_train), weights)
    costs.append(cost_value)
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Cost = {cost_value}")

# Step 7: Evaluate the model on the test set
predictions = np.array([np.sign(quantum_circuit(weights, x)) for x in X_test])

# Accuracy calculation
accuracy = np.mean(predictions == y_test)
print(f"Test Accuracy: {accuracy:.2f}")

# Step 8: Visualize the cost over training epochs
plt.figure(figsize=(8, 6))
plt.plot(costs)
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.title("Cost during Training")
plt.show()

# Step 9: Visualize the decision boundary
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=predictions, cmap='coolwarm', marker='o', label='Predictions')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='coolwarm', marker='x', label='True Labels')
plt.title("Quantum Circuit Classification Results")
plt.legend(loc="best")
plt.show()
```

### ### Quantum Machine Learning - II

```python
# Import necessary libraries
import pennylane as qml
from pennylane import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Step 1: Create a simple synthetic dataset
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)
y = y * 2 - 1  # Convert labels to -1 and 1 for binary classification

# Step 2: Preprocess the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Step 3: Define the quantum circuit (quantum feature map)
n_qubits = 2
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def quantum_circuit(x, weights):
    qml.templates.AngleEmbedding(x, wires=range(n_qubits))
    qml.templates.StronglyEntanglingLayers(weights, wires=range(n_qubits))
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# Step 4: Define the Quantum Support Vector Classifier
def quantum_kernel(x1, x2, weights):
    return np.dot(quantum_circuit(x1, weights), quantum_circuit(x2, weights))

# Step 5: Initialize random weights for the quantum circuit
weights = np.random.randn(3, n_qubits, 3)  # 3 layers, 2 qubits, 3 parameters per layer

# Step 6: Define a classical SVM with quantum kernel (hybrid quantum-classical approach)
from sklearn.svm import SVC

class QuantumSVM:
    def __init__(self, weights):
        self.weights = weights
        self.svc = SVC(kernel=self.quantum_kernel)

    def quantum_kernel(self, X1, X2):
        gram_matrix = np.zeros((len(X1), len(X2)))
        for i, x1 in enumerate(X1):
            for j, x2 in enumerate(X2):
                gram_matrix[i, j] = quantum_kernel(x1, x2, self.weights)
        return gram_matrix

    def fit(self, X, y):
        self.svc.fit(X, y)

    def predict(self, X):
        return self.svc.predict(X)

# Step 7: Train the Quantum SVM
quantum_svm = QuantumSVM(weights)
quantum_svm.fit(X_train, y_train)

# Step 8: Make predictions and evaluate the model
y_pred = quantum_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Quantum SVM Accuracy: {accuracy:.2f}")

# Optional: Visualize the results
import matplotlib.pyplot as plt

plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='coolwarm')
plt.title('Quantum SVM Classification Results')
plt.show()
```
